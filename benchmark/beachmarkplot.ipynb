```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
rtime = pd.read_csv('parallel.csv')
rtime.info()
```

```python
rtime.describe().T
```

```python
plt.subplots(figsize = (30, 7))
sns.set(style = "darkgrid", font_scale = 2.5)
sns.lineplot(x = 'Thread', y = 'hundredTurnsRtime', data = rtime, label = '100 turns', color = 'blue')
sns.lineplot(x = 'Thread', y = 'fourHundredTurnsRtime', data = rtime, label = '400 turns', color = 'orange')
sns.lineplot(x = 'Thread', y = 'sevenHundredTurnsRtime', data = rtime, label = '700 turns', color = 'salmon')
sns.lineplot(x = 'Thread', y = 'thousandTurnsRTime', data = rtime, label = '1000 turns', color = 'peru')
sns.lineplot(x = 'Thread', y = 'fiveThousandTurnsRTime', data = rtime, label = '5000 turns', color = 'gray').set_title('Different Running Time with regards to Threads')
plt.legend(bbox_to_anchor = (1, 1), loc = 2)
```

```python
stime = pd.read_csv('sameRound.csv')
stime.info()
```

```python
stime.describe().T
```

```python
plt.subplots(figsize = (30, 7))
sns.set(style = "darkgrid", font_scale = 2.5)
sns.lineplot(x = 'Thread', y = '16*16', data = stime, label = '16*16', color = 'blue')
sns.lineplot(x = 'Thread', y = '64*64', data = stime, label = '64*64', color = 'orange')
sns.lineplot(x = 'Thread', y = '512*512', data = stime, label = '512*512', color = 'salmon').set_title('Running Time on Different Size with Same Round')
plt.legend(bbox_to_anchor = (1, 1), loc = 2)
```
